```python
plt.figure(figsize=(6,3))
sns.histplot(data=df, x='availability_365')
plt.ylabel('Number of Listings')
plt.title('Availability_365 Distribution')
```

```python
room_types = df['room_type'].unique()  # Get all unique room types

plt.figure(figsize=(8, 5))

for room in room_types:
    sns.histplot(data=df, x='availability_365', hue='room_type', multiple='stack', bins=40)

plt.title('Availability Distribution by Room Type')
plt.xlabel('Availability (days)')
plt.ylabel('Number of Listings')
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
# ['price per bed'] addition of a column
df['price per bed']=df['price']/df['beds']
df.head()
```

```python
#Before 10% hike on revenue
df['revenue_estimate'] = df['price'] * df['availability_365']
df.head()
```

```python
#after 10% hike on revenue
df['increased_revenue'] = df['revenue_estimate'] * 1.10
df.head()
```

```python
sns.barplot(data=df, x='room_type', y='revenue_estimate', color='red', label='Original')
sns.barplot(data=df, x='room_type', y='increased_revenue', color='yellow', label='After 10% Hike', alpha=0.6)
plt.legend()
plt.title('Revenue Before vs After Price Increase by Room Type')
plt.ylabel('Estimated Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.title('Price dependency on neighbourhood')
```

```python
plt.figure(figsize=(8, 5))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df, x='number_of_reviews', y='price')
plt.ylim(0, 2500)
plt.show()
```
